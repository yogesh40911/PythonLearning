```python
def insertionSort(array):
  for i in range(1,len(array)):
    currentNum=array[i]
    #comparing current value with unsorted array
    while i >0 and array[i-1]>currentNum:
      array[i]=array[i-1]
      i=i-1
      array[i]=currentNum


array=[16,23,12,31,12,1,4]
print('Unsorted array')
print(array)


insertionSort(array)
print(' ')
print('Sorted array')
print(array)
```
