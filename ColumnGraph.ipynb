We will learn about a visualization tool namely **Column Graph**, and learn how to create it using Matplotlib. 

A Column Graph is a popular visualization tool. Also known as a Column Chart, it is a type of plot where the length of each bar is proportional to the value of the item that it represents. It is commonly used to compare the values of a variable at a given point in time. 

A column Graph in Python differs from a Bar Graph in its orientation.
While a `column graph is oriented vertically`, a bar graph is oriented horizontally.




```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd

df_europedata=pd.read_csv('https://raw.githubusercontent.com/arpgpt/data-visualization/master/gdp_europe.csv',index_col='country')
years=df_europedata.columns.str.strip('gdpPercapita_')

df_europedata.columns=years.astype(int)
```

```python
df_europedata.plot(kind='bar')
plt.ylabel('GDP per capita')
```

```python
df_europedata.loc['Sweden'].plot(kind='bar')
```

```python
df_europedata.iloc[0:2].plot(kind='bar')
```

```python
df_europedata.iloc[0:1,0:4].plot(kind='bar')
```

```python
df_europedata.loc['Sweden'].plot(kind='bar')
plt.legend(loc='upper left')
```

```python
df_europedata.loc['Sweden'].plot(kind='barh')
```

```python
df_europedata.loc['Sweden'].plot(kind='bar',color='green',edgecolor='red')
```

```python
df_europedata.loc['Sweden'].plot(kind='bar')
plt.title('GDP for sweden')
plt.xlabel('GDP per capital')
plt.ylabel('Years')
plt.xticks(color='red',rotation=45)
plt.yticks(color='green')
```

```python
df_europedata.loc['Sweden'].plot(kind='bar',hatch='*')
```

```python
df_sweden=df_europedata.loc['Sweden']
df_sweden
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.style.use('ggplot')
ax=df_sweden.plot(kind='bar', figsize=(12, 12),width=0.8)
plt.title("GDP Per Capita for Sweden") # add title to the plot
x_offset = -0.3
y_offset = 0.2
for p in ax.patches:
    b = p.get_bbox()
    val = "{:.2f}".format(b.y1 + b.y0)       
    ax.annotate(val, ((b.x0 + b.x1)/2 + x_offset, b.y1 + y_offset))
plt.show()
```
