```python
#Guessing game
import random 
num=random.randint(1,10)

playerName=input("Hello! what's your name")
noOfGuess=0

print('Okay '+ playerName+' I am guessing number between 1 to 10')

while noOfGuess < 5 :
  guess=int(input())
  noOfGuess +=1

  if guess < num:
    print('Your guess is too low.')
  if guess > num:
    print('Your guess is too high')
  if guess==num:
    break

if guess==num:
  print('You guessed the number in '+ str(noOfGuess) + ' tries!')
else:
  print('You did not guess the number was '+ str(num))
```
