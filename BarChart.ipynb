```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd

df_northamerica=pd.read_csv('https://raw.githubusercontent.com/arpgpt/data-visualization/master/north_america.csv', index_col='country')
years = df_northamerica.columns.str.strip("gdpPercap_")

df_northamerica.column = years.astype(int)
```

```python
df_northamerica.plot(kind='bar') #change kind to 'bar', to plot a vertical bar chart
plt.ylabel('GDP per capita')
```

```python
#df_northamerica.loc['Panama'].plot(kind='bar')

df_northamerica.loc['Canada'].plot(kind='bar')
```

```python
#using iloc plotted bar chart
df_northamerica.iloc[0:2].plot(kind='bar')
```

```python
#We may also plot data for specific number of rows and columns using iloc.
df_northamerica.iloc[0:1,0:4].plot(kind='bar')
plt.legend(loc = 4) #Here we used location string
```

```python
#Show Horizontal Bar Chart
df_northamerica.loc['Panama'].plot(kind='barh')
```

```python
#color of bars and borders of bar charts
df_northamerica.loc['Panama'].plot(kind='bar',color='yellow',edgecolor='black')
```

```python
#Add title and axes labels
df_northamerica.loc['Canada'].plot(kind='bar')
plt.title('GDP for canada')
plt.xlabel('GDP per capital')
plt.ylabel('Years')
plt.xticks(color="orange",rotation=45)
plt.yticks(color="red")
```

```python
#Add pattern in bar
df_northamerica.loc['Canada'].plot(kind='bar',hatch="*")
```

```python
#Bar chart with number of observation
df_canada=df_northamerica.loc['Canada']
df_canada
```

```python
df_canada.plot(kind='barh')

for index, value in enumerate(df_canada):
  label = format(int(value),',')

plt.annotate(label, xy=(value,index-0.10),color='black')
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.style.use('ggplot')
ax=df_canada.plot(kind='bar',figsize=(12,12),width=0.8)
plt.title("GDP Per Capita for Canada")
x_offset=-0.3
y_offset=0.2
for p in ax.patches:
  b=p.get_bbox()
  val="{:.2f}".format(b.y1+b.y0)
  ax.annotate(val,((b.x0+b.x1)/2 + x_offset,b.y1+y_offset))

plt.show()
```
