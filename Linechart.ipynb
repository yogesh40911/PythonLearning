We will learn about a visualization tool namely Line Chart, and learn how to create it using Matplotlib.

A line chart or line graph is a type of chart which displays information as a series of data points called ‘markers’ connected by straight line segments. It is a basic type of chart common in many fields.

A line chart is often used to visualize a trend in data over intervals of time – a time series – thus the line is often drawn chronologically. In these cases they are known as run charts.

```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd

time = [0, 1, 2, 3]
position = [0, 100, 200, 300]

plt.plot(time, position)
plt.xlabel('Time (hr)')
plt.ylabel('Position (km)')
plt.annotate('label', xy=(1.0,200), color='black')
```

```python
df_lineeurope = pd.read_csv('https://raw.githubusercontent.com/arpgpt/data-visualization/master/gdp_europe.csv', index_col='country')
years = df_lineeurope.columns.str.strip('gdpPercap_')

df_lineeurope.columns = years.astype(int) # Convert year values to integers, saving results back to dataframe
df_lineeurope.plot()
plt.ylabel('GDP per capita')
```

```python
df_lineeurope.T.plot()
plt.ylabel('GDP per capita')
```

```python
df_lineeurope.loc['Spain'].plot()
df_lineeurope.loc['United Kingdom'].plot()
plt.ylabel('GDP per capita')
plt.legend()
```

```python
df_lineeurope.loc['Spain'].plot(linestyle='-') #solid
df_lineeurope.loc['Norway'].plot(linestyle='--') #dashed
df_lineeurope.loc['Iceland'].plot(linestyle='-.') #dashdot
df_lineeurope.loc['Italy'].plot(linestyle=':') #dotted
df_lineeurope.loc['Sweden'].plot(linewidth=5)
plt.legend()
```
