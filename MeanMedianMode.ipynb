```python
#Numpy short for numeric python is a package that's available by default with Python installation 
#as it is one of the most basic packages we will use along with pandas. 
#We will use the numpy package to calculate mean, median and mode.
```

```python
#calculate mean
import numpy as np
from statistics import mean

list1=[19,29,2,4,22,86,9,55,22,99,1,6]
mean(list1)

np.mean(list1)
```

```python
#calculate median
from statistics import median
median(list1)

np.median(list1)
```

```python
#caculate mode
#Mode is not available in NumPy package. This is because mode is more 
#to do with frequency of occurence of a value than the value itself. 
#So it can hold good for other data types also. As you may recollect, 
#Numpy is only meant for numeric data types.

from statistics import mode
mode(list1)

list2 = ["Bangalore","Gurgaon","Bangalore","Chennai","Mumbai","Mumbai",
         "Delhi","Cochin","Kolkata","Mumbai"]
mode(list2)
```

```python
#Standard deviation
from statistics import stdev
stdev(list1)

np.std(list1)
```
