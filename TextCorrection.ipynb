```python
def pattern(n):
  x=0

  for i in range(0,n):
    x+=1
    for j in range(0,i+1):
      print(x, end=" ")
    
    print('/n')


pattern(5)
```

```python
from textblob import TextBlob
words = ["Data Scence", "Mahine Learnin"]
corrected_words = []
for i in words:
    corrected_words.append(TextBlob(i))
print("Wrong words :", words)
print("Corrected Words are :")
for i in corrected_words:
    print(i.correct(), end=" ")
```
