```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd

df_europedata=pd.read_csv('https://raw.githubusercontent.com/arpgpt/data-visualization/master/gdp_europe.csv', index_col='country')
years= df_europedata.columns.str.strip('gdpPercap_')

df_europedata.column=years.astype(int)
```

```python
df_europedata.plot(kind='bar')
plt.ylabel('GDP per capita')
```

```python
df_europedata.loc['Italy'].plot(kind='bar')
```

```python
#using iloc plotted bar chart
df_europedata.iloc[0:2].plot(kind='bar')
```

```python
df_europedata.iloc[0:1,0:7].plot(kind='bar')
plt.legend(loc=4)
```

```python
df_europedata.loc['Norway'].plot(kind='barh')
```

```python
df_europedata.loc['Norway'].plot(kind='bar',color='green',edgecolor='black')
```

```python
df_europedata.loc['Norway'].plot(kind='bar')
plt.title('GDP for Norway')
plt.xlabel('GDP per capital')
plt.ylabel('Years')
plt.xticks(color='orange',rotation=90)
plt.yticks(color='red')
```

```python
df_europedata.loc['Norway'].plot(kind='bar',hatch='*')
```

```python
df_norway=df_europedata.loc['Norway']
df_norway
```

```python
df_norway.plot(kind='barh')

for index, value in enumerate(df_norway):
  label=format(int(value),',')

plt.annotate(label, xy=(value,index-0.10),color='red')
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.style.use('ggplot')
ax=df_norway.plot(kind='bar',figsize=(12,12),width=0.8)
plt.title("GDP Per Capita for Norway")
x_offset=-0.3
y_offset=0.2
for p in ax.patches:
  b=p.get_bbox()
  val="{:.2f}".format(b.y1+b.y0)
  ax.annotate(val,((b.x0+b.x1)/2 + x_offset,b.y1+y_offset))

plt.show()
```
